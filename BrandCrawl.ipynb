```python
#Gọi các thư viện
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#Mở Chrome driver và mở trang web
driver = webdriver.Chrome()
driver.get('https://www.agencyvietnam.com/#')
```

```python
#Lấy ra toàn bộ menu và đường link từng menu
soup = BeautifulSoup(driver.page_source)
menu = soup.find_all('a',class_ = 'menu-link')
menu_lst = []
for i in menu[2:-1]:
    menu_lst.append([i['href'],i.text[1:-1]])
```

```python
menu_lst
```

```python
#Tạo list rỗng để chứa các giá trị
cat1 = []
cat2 = []
link = []

#Lặp qua từng đường link menu để lấy ra segment
for me in menu_lst:
    driver.get(me[0])
    soup = BeautifulSoup(driver.page_source)
    raw = soup.find_all('div',class_='owl-item')

    #Lặp qua từng segment để lấy link công ty
    for i in raw:
        sub_title = i.find_all('h3', class_ = 'd-inline-block text-center p-2 w-100 text-truncate')[0].text
        brand_links = i.find_all('a')
        for brand in brand_links:
            if brand['href'] != '#':
                link.append('https://www.agencyvietnam.com'+brand['href'])
                cat1.append(me[1])
                cat2.append(sub_title)
```

```python
df = pd.DataFrame({
                    'menu':cat1,
                    'segment': cat2,
                    'link':link,
                  },columns = ['menu','segment','link'])
```

```python
df
```

```python
#Tạo list rỗng để chứa các giá trị
name = []
mail = []
phone = []
adress = []
web = []
intro = []
personel = []
discipline = []
group = []
present = []
position = []

#Lặp qua từng đường link công ty để lấy về thông tin, nếu không có sẽ trả về ""
for i in range(0,len(df)):
    driver.get(df.iloc[i]['link'])
    print(df.iloc[i]['link'])
    name.append(driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div/div[2]/h1').text)

    #Mail
    try:
        mail.append(driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div/div[2]/div/div[3]/a').text)
    except:
        mail.append('')

    #Sdt
    try:
        phone.append(driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div/div[2]/div/div[2]').text)
    except:
        mail.append('')
    
    #Địa chỉ
    try:
        adress.append(driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div/div[2]/div/div[1]').text)
    except:
        adress.append('')
    
    #Trang web
    try:
        web.append(driver.find_element(By.XPATH,'/html/body/main/div/div[1]/div/div/div[2]/div/div[4]/a').text)
    except:
        web.append('')
    
    #Giới thiệu
    try:
        intro.append(driver.find_element(By.XPATH,'/html/body/main/div/div[2]/div[1]/p[2]').text)
    except:
        intro.append('')
    
    #Discipline
    try:
        discipline.append(driver.find_element(By.XPATH,'/html/body/main/div/div[2]/div[2]/div/div/span[2]/span').text)
    except:
        discipline.append('')
    
    #Tập đoàn/ tổ chức
    try:
        group.append(driver.find_element(By.XPATH,'/html/body/main/div/div[2]/div[2]/div/div/span[3]/span').text)
    except:
        group.append('')
    
    #Link bài present
    try:
        #vì find_element không lấy được thông tin nằm trong thẻ tags <> nên phải sử dụng BeautifulSoup
        soup = BeautifulSoup(driver.page_source)
        y = soup.find_all('div',class_ = 'embed-responsive embed-responsive-16by9 my-20')
        if len(y) > 0:
            x = str(y[0])
            present.append(x[x.find('src')+7:-18])
        else:
            present.append('')
    except:
        present.append('')
    
```

```python
#Ghép vào dataframe
df['name'] = name
df['mail'] = mail
df['phone'] = phone
df['adress'] = adress
df['web'] = web
df['intro'] = intro
df['discipline'] = [i.split('\n')[1] if i != '' else '' for i in discipline]
df['present'] = present
df['group'] = [i.split('\n')[1] if i != '' else '' for i in group]
```
